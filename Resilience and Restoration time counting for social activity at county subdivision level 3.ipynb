# the codes below from here is to find the intersection of baseline and activity line

```python
x = [(0,0), (3,4), (5,6)]
y = [(3,4), (5,6), (5,6)]
```

```python
def line_intersection(line1, line2):
    xdiff = (line1[0][0] - line1[1][0], line2[0][0] - line2[1][0])
    ydiff = (line1[0][1] - line1[1][1], line2[0][1] - line2[1][1])

    def det(a, b):
        return a[0] * b[1] - a[1] * b[0]

    div = det(xdiff, ydiff)

    if div != 0:
        d = (det(*line1), det(*line2))
        x = det(d, xdiff) / div
        y = det(d, ydiff) / div
        return x, y
    else:
        print("lines do not intersect")
       #continue
        
    #return x, y
```

```python
ints = []
for (a, b) in zip(x, y):
    print(a,b)
    ints.append(line_intersection((a, b), ((1,0), (0,1))))
```
